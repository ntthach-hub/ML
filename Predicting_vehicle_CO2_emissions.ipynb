#### Features

**Make:** Thương hiệu của xe.

**Model:** Mẫu xe.

**Vehicle Class:** Phân loại xe (ví dụ: xe compact, SUV).

**Engine Size (L):** Dung tích động cơ tính bằng lít.

**Cylinders:** Số xi-lanh trong động cơ.

**Transmission:** Loại hộp số (ví dụ: tự động, số tay).

**Fuel Type:** Loại nhiên liệu sử dụng (ví dụ: xăng, diesel).

**Fuel Consumption City (L/100 km):** Mức tiêu thụ nhiên liệu trong thành phố (lít trên 100 km).

**Fuel Consumption Hwy (L/100 km):** Mức tiêu thụ nhiên liệu khi chạy đường cao tốc (ngoài thành phố).

**Fuel Consumption Comb (L/100 km):** Mức tiêu thụ nhiên liệu kết hợp (thành phố và cao tốc).

**Fuel Consumption Comb (mpg):** Mức tiêu thụ nhiên liệu kết hợp tính theo dặm trên mỗi gallon.

**CO2 Emissions (g/km):** Lượng khí CO2 thải ra tính theo gam trên mỗi km.


Ứng dụng của bộ dữ liệu: Bộ dữ liệu này có thể được sử dụng để hiểu rõ hơn về hiệu suất tiêu thụ nhiên liệu và tác động môi trường của các loại xe. # Simple Linear Regression Model (Mô hình hồi quy tuyến tính bội)

## Chia tập dữ liệu thành X (biến độc lập) và y (biến phụ thuộc)

Đầu tiên, lựa chọn biến đầu vào là dung tích động cơ (Engine Size) và biến mục tiêu là lượng khí thải CO2 (CO2 Emissions). Sau đó, dữ liệu được chia thành hai tập: tập huấn luyện (80%) và tập kiểm tra (20%) để đảm bảo mô hình được đánh giá khách quan.
Biến đầu vào (X): **Engine Size (L)**
Biến mục tiêu (y): **CO2 Emissions (g/km)**

```python
#  Lựa chọn biến đầu vào và biến mục tiêu
X = data[["Engine Size(L)"]]
y = data["CO2 Emissions(g/km)"]
```

### training data

```python
# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 101)
```

```python
# Hiển thị kích thước (shape) của các tập dữ liệu sau khi xử lý

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

**Kết quả:** Số lượng mẫu tập huấn luyện: 5025
	Số lượng mẫu tập kiểm tra: 1257

### Training the Model

Sau khi chia dữ liệu, nhóm sử dụng mô hình hồi quy tuyến tính đơn để huấn luyện trên tập train. Mô hình được xây dựng bằng thư viện scikit-learn. Sau khi huấn luyện, mô hình được sử dụng để dự đoán lượng khí thải CO2 trên tập kiểm tra

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

### Predicting Test Data

```python
y_pred = model.predict(X_test)
model.coef_
```

```python
model.intercept_
```

### Model evaluation

Đánh giá mô hình

```python
# Dự đoán trên dữ liệu huấn luyện
y_train_pred = model.predict(X_train)
```

```python
# Khối mã dùng để tính toán và ghi lại các chỉ số sai số
def train_val(y_train, y_train_pred, y_test, y_pred, i): 
    
    scores = {
    i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}
    }
    
    return pd.DataFrame(scores)
```

```python
slr = train_val(y_train, y_train_pred, y_test, y_pred, 'linear')
slr
```

```python
avg_co2= data['CO2 Emissions(g/km)'].mean()
avg_co2
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

Các điểm thực tế (chấm xanh) phân bố quanh đường dự đoán, cho thấy mô hình đã học được xu hướng tổng thể.
Tuy nhiên, mô hình chưa dự đoán tốt với các trường hợp ngoại lệ hoặc giá trị Engine Size lớn.
Kết quả này phù hợp với bản chất của hồi quy tuyến tính đơn, chỉ sử dụng một biến đầu vào nên chưa thể hiện hết sự phức tạp của dữ liệu thực tế.

```python
# linear model and visualizer

model = LinearRegression()
visualizer = ResidualsPlot(model)
visualizer.fit(X_train, y_train)  # Huấn luyện mô hình với dữ liệu huấn luyện và truyền vào công cụ trực quan hóa
visualizer.score(X_test, y_test)  # Đánh giá mô hình trên tập dữ liệu kiểm tra
visualizer.show();                 # Hiển thị và hoàn tất biểu đồ
```

## Predicting test data

## Đánh giá mô hình

```python
y_true = y_test
y_pred = model.predict(X_test)

mae  = round(mean_absolute_error(y_true, y_pred), 2)
rmse = round(np.sqrt(mean_squared_error(y_true, y_pred)), 2)
r2   = round(r2_score(y_true, y_pred), 3)

print(f"MAE:  {mae} g/km")
print(f"RMSE: {rmse} g/km")
print(f"R²:   {r2}")
```

```python
def train_val(y_train, y_train_pred, y_test, y_pred, i):
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
              
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
mlr = train_val(y_train, y_train_pred, y_test, y_pred, "linear")
mlr
```

Mô hình hồi quy tuyến tính bội cho kết quả R2 trên tập test là 0.869, RMSE khoảng 21. Điều này cho thấy mô hình dự đoán khá tốt, sai số trung bình nhỏ.

```python
plt.figure(figsize=(6,6))
plt.scatter(y_true, y_pred, alpha=0.5)
plt.plot([y_true.min(), y_true.max()],
         [y_true.min(), y_true.max()],
         'r--', lw=2)
plt.xlabel("CO2 thực tế (g/km)")
plt.ylabel("CO2 dự đoán (g/km)")
plt.title("Thực tế vs Dự đoán")
plt.tight_layout()
plt.show()
```

Một số điểm lệch xa khỏi đường chéo, điều này thể hiện mô hình dự đoán kém chính xác ở một vài trường hợp ngoại lệ. Phân bố điểm không có dấu hiệu lệch hệ thống (bias), điều này hỗ trợ giả định rằng mô hình không bị lệch theo một chiều cố định.

```python
residuals = y_true - y_pred

plt.figure(figsize=(6,4))
plt.hist(residuals, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel("Sai số (g/km)")
plt.ylabel("Số lượng mẫu")
plt.title("Phân phối sai số")
plt.tight_layout()
plt.show()
```

```python
# Cross Validate (CV)
model = LinearRegression()

scores = cross_validate(estimator=model, X= X_train, y=y_train,
               scoring= ["r2", "neg_mean_absolute_error", "neg_mean_squared_error", "neg_root_mean_squared_error"],
               cv=10,
               return_train_score=True
              )
scores = pd.DataFrame(scores, index= range(1,11))
scores.iloc[:, 2:].mean()
```

```python
# X_train, y_train đã có sẵn
models = {
    'Linear': LinearRegression(),
    'Ridge':   Ridge(alpha=1.0),
    'Lasso':   Lasso(alpha=0.1),
    'Elastic': ElasticNet(alpha=0.1, l1_ratio=0.5)
}
```

```python
# —— Cross-Validation Scores ——
cv_results = []
for name, model in models.items():
    # Dùng 5-fold CV với scoring là RMSE và R2
    # scoring ‘neg_root_mean_squared_error’ trả về -RMSE
    neg_rmse_scores = cross_val_score(model, X_train, y_train,
                                      cv=5,
                                      scoring='neg_root_mean_squared_error')
    r2_scores      = cross_val_score(model, X_train, y_train,
                                     cv=5,
                                     scoring='r2')
    rmse_scores = -neg_rmse_scores  # chuyển thành dương
    cv_results.append({
        'Model': name,
        'RMSE Mean': rmse_scores.mean(),
        'RMSE Std':  rmse_scores.std(),
        'R2 Mean':   r2_scores.mean(),
        'R2 Std':    r2_scores.std()
    })

cv_df = pd.DataFrame(cv_results)
print("Cross-Validation Results:")
print(cv_df.round(3))
```

```python
def cv_report(model, X, y, name, cv=5):
    """Chạy CV, in ra số và vẽ biểu đồ cho từng model."""
    # RMSE (neg_root_mean_squared_error trả về -RMSE)
    neg_rmse = cross_val_score(model, X, y,
                               cv=cv,
                               scoring='neg_root_mean_squared_error')
    rmse = -neg_rmse
    r2    = cross_val_score(model, X, y, cv=cv, scoring='r2')

    # In kết quả
    print(f"--- CV {cv}-fold cho {name} ---")
    print(f"RMSE trung bình = {rmse.mean():.3f} ± {rmse.std():.3f}")
    print(f"R²   trung bình = {r2.mean():.3f} ± {r2.std():.3f}\n")

    # Trả về DataFrame để bạn có thể gom vào bảng chung nếu muốn
    return pd.DataFrame([{
        'Model': name,
        'RMSE Mean': rmse.mean(),
        'RMSE Std':  rmse.std(),
        'R2 Mean':   r2.mean(),
        'R2 Std':    r2.std()
    }])

def plot_cv_bar(cv_df, metric):
    """Vẽ bar chart cho metric='RMSE Mean' or 'R2 Mean'."""
    plt.figure(figsize=(6,4))
    plt.bar(cv_df['Model'], cv_df[metric])
    plt.title(f'Hình: So sánh {metric} qua {len(cv_df)} mô hình')
    plt.xlabel('Mô hình')
    plt.ylabel(metric)
    plt.tight_layout()
    plt.show()



# —— 5.1.2 Cross-Validation cho Linear Regression ——
cv_lin = cv_report(models['Linear'], X_train, y_train, name='Linear Regression')



plot_cv_bar(cv_lin, 'RMSE Mean')
plot_cv_bar(cv_lin, 'R2 Mean')
```

```python
cv_list = []
for name, model in models.items():
    df = cv_report(model, X_train, y_train, name=name)
    cv_list.append(df)

cv_df = pd.concat(cv_list, ignore_index=True)
# Hiển thị bảng
print("Bảng 5.x: Kết quả 5-fold CV cho tất cả mô hình")
print(cv_df.round(3).to_markdown(index=False))

# Vẽ biểu đồ chung
plot_cv_bar(cv_df, 'RMSE Mean')
plot_cv_bar(cv_df, 'R2 Mean')
```
